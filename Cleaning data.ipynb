### 3. Cleaning missing values

```python
stumps = tree_census_subset[tree_census_subset['status'] == 'Stump']
stumps
```

```python
deads = tree_census_subset[tree_census_subset['status'] == 'Dead']
deads
```

```python
mask = ((tree_census_subset['status'] == 'Dead') | (tree_census_subset['status'] == 'Stump'))
```

```python
#Replace NaN values in all columns with 'Not Applicable' for: Rows of Stump and Dead in 'status' 
#.loc(row_indexer, col_indexer) => Specific column: .col(mask, 'health')
tree_census_subset.loc[mask] = tree_census_subset.loc[mask]. fillna('Not Applicable')
```

```python
tree_census_subset [tree_census_subset['status'] == 'Stump']
```

```python
tree_census_subset [tree_census_subset['status'] == 'Dead']
```

```python
#Check any other missing values left
tree_census_subset.isna().sum()
```

```python
#Show NaN 'steward' values: Check for problems, health, sidewalk, spc_latin as well
tree_census_subset[tree_census_subset['steward'].isna()]
```

```python
#Print steward values & total counts of each value type
#Check for problems, health, sidewalk, spc_latin as well
tree_census_subset ['steward'].value_counts()
```

```python
#Replace missing values
tree_census_subset ['problems'].fillna('None', inplace=True)
tree_census_subset ['health'].fillna('Good', inplace=True)
tree_census_subset ['spc_latin'].fillna('No Observation', inplace=True)
tree_census_subset ['sidewalk'].fillna('NoDamage', inplace=True)
tree_census_subset ['steward'].fillna('None', inplace=True)
```

```python
#Last check of missing values
tree_census_subset.isna().sum()
```

```python

tree_census_subset_alive = tree_census_subset[tree_census_subset['status'] == 'Alive']
tree_census_subset_dead_or_stump = tree_census_subset[(tree_census_subset['status'] == 'Dead') | (tree_census_subset['status'] == 'Stump')]
```

```python
tree_census_subset_alive.groupby('spc_latin')['tree_dbh'].describe()
```

```python
#Abnormal value (0) from the min of tree diameter
tree_census_subset_alive[tree_census_subset_alive['tree_dbh']==0]
```

```python
#Reference Table:
stats_alive = tree_census_subset_alive.groupby('spc_latin')['tree_dbh'].describe().reset_index()[['spc_latin', '25%', '75%']]
stats_alive
```

```python

tree_census_subset_alive.merge(stats_alive, on = 'spc_latin', how = 'left')
```

```python
#Shrink the values that are higher than 75% percentile value of the tree
#And increase the values lower than 25% percentile value of the tree
#mask -> if/then idiom

mask = tree_census_subset_alive['tree_dbh'] < tree_census_subset_alive ['25%']
tree_census_subset_alive.loc[mask,'tree_dbh'] = tree_census_subset_alive ['25%']

mask = tree_census_subset_alive['tree_dbh'] > tree_census_subset_alive['75%']
tree_census_subset_alive.loc[mask,'tree_dbh'] = tree_census_subset_alive ['75%']
```
